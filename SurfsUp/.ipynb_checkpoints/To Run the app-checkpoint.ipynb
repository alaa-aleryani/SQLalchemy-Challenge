```python
# Import the dependencies.
import pandas as pd
import numpy as np
import datetime as dt

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func

# Import the flask dependency # flask: as an api - to transport info
from flask import Flask, jsonify

#################################################
# Database Setup
#################################################
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(autoload_with=engine)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station

# Create our session (link) from Python to the DB
session = Session(engine)

#################################################
# Flask Setup
#################################################
app = Flask(__name__)  # instantiate the flask library

```

```python
#################################################
# Flask Routes
#################################################
@app.route("/")

def welcome():
    """ List all available api routes. """
    return (
        f"Welcome To Our Homepage!<br/>"
        f"Available Routes:<br/>"
        f"/api/v1.0/precipitation<br/>"
        f"/api/v1.0/stations<br/>"
        f"/api/v1.0/tobs"
    )

# --------------------------------------------------
# Precipitation Route
@app.route("/api/v1.0/precipitation")

# do not run this cell twice otherwise
# if I get an error change the name of the function maybe add s
# or restart the kernel
def percipitation():  
    
    ## Returns json with the date as the key and the value as the precipitation   
    # Rain for the last year
    
#     # get the latest date
#     recent_date_str = session.query(Measurement.date).\
#         order_by(Measurement.date.desc()).first()[0]

    # get the date one year from the latest date
    global prev_year_date
    prev_year_date = dt.date(2017,8,23) - dt.timedelta(days=365)  # otherwise just pass the actual date
    
    # query the data
    percip_query = session.query(Measurement.date, Measurement.prcp).\
        filter(Measurement.date >= prev_year_date).all()
    
    # session.close
    session.close()
    
    # the data we want
    percip = {date:prcp for date, prcp in percip_query}  # list comprehension
    return jsonify(percip)  # object variable inside the function
    print(type(percip))

# -------------------------------------------------------------------------
## Stations Route
# Return a JSON list of stations from the dataset.
@app.route("/api/v1.0/stations")           # START UNCOMMENTING HERE

def stations():
        
    # query the data
    station_query = session.query(Station.station).all()

    # session.close
    session.close()

    # the data we want
#     stations_list = [station_id for station_id in station_query]

    # Convert list of tuples into normal list
    stations_list = list(np.ravel(station_query))
    return jsonify(stations_list)
    print(type(station_query))
# -------------------------------------------------------------------------
# # Tobs Route
# Query the dates and temperature observations of the most-active station for the previous year of data.
# Return a JSON list of temperature observations for the previous year.
@app.route("/api/v1.0/tobs")           

def tobs():
    
    # query the data
    temp_obs_data = session.query(Measurement.tobs).\
        filter(Measurement.station == 'USC00519281').\
        filter(Measurement.date >= prev_year_date).all()

    # session.close()
    session.close()
    
    # the data we want
#     temp_obs_data_list = [temp for temp in temp_obs_data_list]

    # Convert list of tuples into normal list
    temp_obs_data_list = list(np.ravel(temp_obs_data))
    return jsonify(temp_obs_data_list)
```
